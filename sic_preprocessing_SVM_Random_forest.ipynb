# Data Preprocessing

```python
# Separating the date from the time in different columns

combined['trans_date_trans_time'] = pd.to_datetime(combined['trans_date_trans_time'])
# Create separate columns for date and time
combined['date'] = combined['trans_date_trans_time'].dt.date

combined['time'] = combined['trans_date_trans_time'].dt.time
```

```python
# Calculatign the age using the DOB and transaction date
combined['dob'] = pd.to_datetime(combined['dob'])
combined['age'] = (combined['trans_date_trans_time'] - combined['dob']).dt.days // 365

```

```python
combined['trans_hour'] = pd.to_datetime(combined['trans_date_trans_time']).dt.hour
```

```python
# Removing Redundant columns such as identifiers
combined = combined.drop(columns=['Unnamed: 0','cc_num','merchant','trans_num','unix_time','first',
                                  'last','street','zip' , 'trans_date_trans_time' , 'date' , 'dob' , 'lat' , 'long' , 'merch_lat' , 'merch_long','time'  ])
# Date was removed but time was kept since there might be a correlation between frauds and time
# City was removed sicne we have the state and due to having over 800 cities which might be a problem in creating dummies
```

```python
combined.head()
```

```python
print(combined.dtypes)
```

```python
combined.head(10)
```

```python
# Change datatype of is_fraud to string
combined['is_fraud'] = combined['is_fraud'].astype(int)
combined['category'].value_counts()
```



*   **We have more than 14 categories, creating dummy variables for them would be alot, hence, we need to combine similarities**




```python
combined['category'].unique()
```

```python
# Creating processed_df to work on instead of the same combined df
processed_df = combined
```

```python
processed_df['category'] = processed_df['category'].replace(['gas_transport', 'travel'], 'transportation')
processed_df['category'] = processed_df['category'].replace(['home',  'personal_care', 'kids_pets'], 'household')
processed_df['category'] = processed_df['category'].replace(['shopping_pos', 'shopping_net' , 'grocery_pos' , 'grocery_net'], 'shopping&grocery')
processed_df['category'] = processed_df['category'].replace(['entertainment', 'misc_pos' ,'misc_net' ], 'entertainment&others')
processed_df['category'] = processed_df['category'].replace(['food_dining', 'health_fitness'], 'food&health')
```



*   **Now we have only 5 categories and will end up with additional 4 dummy variables instead of 13**




```python
processed_df['city'].value_counts()
```

```python
processed_df['state'].value_counts()
```

```python
processed_df['job'].value_counts()
```

```python
processed_df = processed_df.drop(columns=['job' ,  'city' ])
```

```python
processed_df.head()
```

###**Encoding**

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
processed_df['gender'] = le.fit_transform(processed_df['gender'])

# 1 for males and 0 for females

processed_df.head()
```

```python
# Generate dummy variables for 'category' and 'state' columns
category_dummies = pd.get_dummies(processed_df['category'], prefix='category', drop_first=True)
state_dummies = pd.get_dummies(processed_df['state'], prefix='state', drop_first=True)

# Convert dummies to integer type instead of boolean
category_dummies = category_dummies.astype(int)
state_dummies = state_dummies.astype(int)

# Concatenate the original DataFrame with the dummies DataFrames
processed_df = pd.concat([processed_df, category_dummies, state_dummies], axis=1)

# Drop the original 'category' and 'state' columns if you no longer need them
processed_df = processed_df.drop(columns=['category', 'state'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit and transform the specified columns
processed_df[['amt' , 'city_pop' ,'age' , 'trans_hour']] = scaler.fit_transform(processed_df[['amt' , 'city_pop','age' , 'trans_hour']])
```

# Random Forest

```python
processed_df = pd.read_csv('/content/processed_data.csv')
```

```python
# Creating a separate df for this model in case of modifications

randomforest_df = processed_df
```

### Separating the features and the label

```python
randomforest_df['is_fraud'].value_counts()
```



*   **Most of the transactions were legitimate with around 0.5% of the transactions were fraudulent**






*   **Randomforest only works with either strings or floats**




```python
# Separating the features from the label

X = randomforest_df.drop('is_fraud', axis=1).values
y = randomforest_df['is_fraud'].values
```

```python
print(f'features.shape: {X.shape}')
print(f'label.shape: {y.shape}')
```

### Splitting data to Train-Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.3, random_state = 0)
```



*   **Using stratify=y so that the data will be split in such a way that the proportion of each class in y will be approximately the same in both the training and test sets**
*   **This argument is useful for imbalanced datasets**
*   **random_state = 0 ensures that the split is consistent across multiple runs of the code**




```python
print(f'features_train.shape: {X_train.shape}')
print(f'label_train.shape: {y_train.shape}')
print(f'features_train_test.shape: {X_test.shape}')
print(f'label_train_test.shape: {y_test.shape}')
```



*   **Finding the most suitable hyper parameter using Random Search**




```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python


# Define the parameter distribution
param_dist = {
    'max_depth': [3, 5, 7, 10, 15],
    'n_estimators': randint(5, 100),  # Randomly sample between 5 and 100
    'criterion': ['gini', 'entropy']
}

# Initialize the RandomForestClassifier
rf = RandomForestClassifier()

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, n_iter=20, cv=5, scoring='accuracy', random_state=0)

# Fit RandomizedSearchCV
random_search.fit(X_train, y_train)

# Print the best parameters and score
print("Best parameters found:")
print(random_search.best_params_)
print("Best score achieved:")
print(random_search.best_score_)
```

```python
import matplotlib.pyplot as plt

depths = [3, 5, 7, 10, 15]
train_scores = []
test_scores = []

for depth in depths:
    rf = RandomForestClassifier(max_depth=depth, n_estimators=47)
    rf.fit(X_train, y_train)
    train_scores.append(rf.score(X_train, y_train))
    test_scores.append(rf.score(X_test, y_test))

plt.plot(depths, train_scores, label='Training Score')
plt.plot(depths, test_scores, label='Test Score')
plt.xlabel('Max Depth')
plt.ylabel('Score')
plt.title('Learning Curves for Different Tree Depths')
plt.legend()
plt.show()
```



*   **We can see that the suitable depth for each tree is 10 and the number of trees is 47**
*   **There's a marginal contribution to the score for the depth from 10 to 15, we will choose depth of 10 for lower complexity**




```python
model = RandomForestClassifier(max_depth = 10, n_estimators = 47,criterion = 'entropy' ,
                               random_state=42 , class_weight='balanced' ,
                               oob_score=True)
model.fit(X_train, y_train)
```



*   **class_weight controls the weights associated with classes to handle imbalanced datasets**
*   **class_weight='balanced' ensures that the model doesn't overly favor the majority class by giving the majority class lower weights and the minority higher weights**



```python
columns = list(randomforest_df.columns)
columns
```

```python
del columns[3]
```

```python
#feature importance random forest
feature_imp = pd.Series(model.feature_importances_,index=columns).sort_values(ascending=False)
type(feature_imp)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sort feature importance in descending order and select the top 9
top_n = 9
sorted_indices = feature_imp.sort_values(ascending=False)[:top_n]

# Visualize feature importance for the top 9 features
plt.figure(figsize=(10, 6))  # Adjust the figure size for clarity
sns.barplot(x=sorted_indices, y=sorted_indices.index)

# Add labels and title to the graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Top 9 Important Features")
plt.show()
```



*   **The plot above shows that the most important feature in determining fradulent transactions is the transaction's amount**




### Evaluating the model

```python
from sklearn.metrics import classification_report

# Making predictions on the test set
y_pred = model.predict(X_test)

# Generate the classification report
report = classification_report(y_test, y_pred)
print(report)
```



*   **recall for 1 wasn't satisfactory since it was 0.46**
*   **increasing the weight for the minority class improced the recall to 0.82**




```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
print(f'Accuracy: {accuracy_score(y_test, y_pred) * 100}%')
```

```python
from sklearn.metrics import roc_curve, auc

probs = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

### Evaluating the model

```python
# Separating the features from the label

X = svm_df.drop('is_fraud', axis=1).values
y = svm_df['is_fraud'].values
```

```python
print(f'features.shape: {X.shape}')
print(f'label.shape: {y.shape}')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.svm import SVC


model = SVC(kernel='rbf' , class_weight='balanced' , random_state = 1)
model.fit(X_train, y_train)
```

```python
percentage = model.score(X_test, y_test)
print(f"Accuracy = {percentage*100} %")
```

```python
# Making predictions on the test set
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns

y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
print(f'Accuracy: {accuracy_score(y_test, y_pred) * 100}%')
```

```python
# Generate the classification report

from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import roc_curve, auc

# Calculate decision scores for the positive class
decision_scores = model.decision_function(X_test)

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, decision_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
